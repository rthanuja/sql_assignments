```python
#Q-13. Write an SQL query to show only odd rows from aÂ table.
query="select * from (SELECT *,ROW_NUMBER() OVER (ORDER BY (SELECT NULL)) AS RowNum FROM worker) as new where rownum%2<>0; ;"
df=pd.read_sql(query,engine)
df
```

```python
#Q-14. Write an SQL query to clone a new table from another table.
query="create table if not exists worker_clone as select * from worker;"
engine.execute(query)
df=pd.read_sql("show tables",engine)
df
```
