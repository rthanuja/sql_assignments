```python
import pandas as pd
from sqlalchemy import create_engine
connection=create_engine('mysql+mysqlconnector://root:Mysql@localhost:3306/Ineuron')
engine=connection.connect()
```

```python
#Q-17. Write an SQL query to show the top n (say 10) records of a table.
n=int(input("enter the number"))
query=f"select * from worker limit {n}";
df=pd.read_sql(query,engine)
df
```

```python
#Q-18. Write an SQL query to determine the nth (say n=5) highest salary from a table.
n=int(input("enter the number"))
query=f"select distinct salary from worker order by salary desc limit 1 offset {n-1}";
df=pd.read_sql(query,engine)
df
```
